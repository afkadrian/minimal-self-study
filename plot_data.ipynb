```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Data organization
categories = ['Standard', 'Proprioception', 'Direct Motion', 'Developmental Motion']
groups = ['Group 1', 'Group 2']
conditions = ['Full', 'Partial', 'None']

data = {
    'Standard': {
        'Group 1': (3, 4, 3),
        'Group 2': (2, 2, 6)
    },
    'Proprioception': {
        'Group 1': (5, 3, 2),
        'Group 2': (1, 2, 7)
    },
    'Direct Motion': {
        'Group 1': (6, 1, 3),
        'Group 2': (5, 1, 4)
    },
    'Developmental Motion': {
        'Group 1': (3, 0, 7),
        'Group 2': (1, 0, 9)
    }
}

# Setup plot
plt.figure(figsize=(14, 8))
sns.set_style("whitegrid")
palette1 = ['#2F4F4F', '#4682B4', '#87CEEB']
palette2 = ["#541D1D", "#94533B", "#EBAD87"]

# Calculate positions
n_categories = len(categories)
x = np.arange(n_categories)  # Category positions
width = 0.35  # Width of each group bar

# Plot each group's stacked bars
for group_idx, group in enumerate(groups):
    bottom = np.zeros(n_categories)  # Initialize bottom positions
    palette = palette1 if group == "Group 1" else palette2
    # Plot each condition segment
    for cond_idx in range(len(conditions)):
        values = np.array([data[cat][group][cond_idx] for cat in categories])
        
        # Calculate x positions: group0 on left, group1 on right
        x_pos = x + (width * group_idx)
        
        plt.bar(
            x_pos, 
            values, 
            width, 
            bottom=bottom,
            color=palette[cond_idx],
            edgecolor='white',
            label=conditions[cond_idx] if group_idx == 0 else None
        )
        bottom += values

# Finalize plot
#plt.title('Grouped and Stacked Comparison by Category', fontsize=16, pad=15)
plt.xlabel('Extension', fontsize=12)
plt.ylabel('Trials', fontsize=12)
plt.xticks(x + width/2, categories, fontsize=11)  # Center labels between groups

# Create compact 3-row legend
legend_handles = []
legend_labels = []

for cond_idx in range(3):
    # Add Group 1 entry
    legend_handles.append(plt.Rectangle((0,0), 1, 1, color=palette1[cond_idx]))
    legend_labels.append(f'{conditions[cond_idx]} (Group 1)')
for cond_idx in range(3): 
    # Add Group 2 entry
    legend_handles.append(plt.Rectangle((0,0), 1, 1, color=palette2[cond_idx]))
    legend_labels.append(f'{conditions[cond_idx]} (Group 2)')

# Create legend with 2 columns and 3 rows
plt.legend(
    legend_handles,
    legend_labels,
    loc='upper center',
    bbox_to_anchor=(0.5, 1.15),  # Position above the plot
    ncol=2,
    frameon=True,
    title='Level of Control Verified'
)


# Add value labels to segments
for cat_idx, category in enumerate(categories):
    for group_idx, group in enumerate(groups):
        bottom = 0
        x_pos = cat_idx + (width * group_idx)
        
        for cond_idx, value in enumerate(data[category][group]):
            if value > 0:  # Only label non-zero segments
                y_pos = bottom + value/2
                plt.text(
                    x_pos, 
                    y_pos, 
                    str(value), 
                    ha='center', 
                    va='center',
                    fontsize=9,
                    fontweight='bold',
                    color='white'
                )
            bottom += value

# Add group labels at top
for group_idx, group in enumerate(groups):
    x_pos = x + (width * group_idx)
    totals = [sum(data[cat][group]) for cat in categories]
    
    for cat_idx, total in enumerate(totals):
        plt.text(
            x_pos[cat_idx], 
            total + 0.2,  # Slightly above bar
            group, 
            ha='center', 
            va='bottom',
            fontsize=10,
            fontweight='bold'
        )

plt.tight_layout()
plt.show()
```
